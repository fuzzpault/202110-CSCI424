```
class Gradebook:
  # The best way to set and initialize variables is in the __init__ function (constructor)
  def __init__(self):
    print("init called")
    self.students = {}
  #students = {} # Another way, but not recommended
  def addStudent(self, student_name):
    self.students[student_name] = []
    
  def addGrade(self, student_name, grade):
    if student_name not in self.students.keys():
      print("{} is not a student, adding them now".format(student_name))
      #self.students[student_name] = []
      self.addStudent(student_name)
    self.students[student_name].append(grade)
    
  def printAll(self):
    print("Here are the {} students in the gradebook:".format(len(self.students.keys())))
    for name in self.students.keys():
      print("{}: {}".format(name, self.students[name]))
```

```
gb = Gradebook()
gb.addStudent("Paul")
gb.addStudent("Hassan")
gb.addGrade("Paul", 75)
gb.addGrade("Hassan", 76)
gb.addGrade("Bob", 76)
gb.printAll()
```

```
# Demo of global variables
# By declaring a variable global, you say to use the global variable rather than making a local one
def stuff():
  global gb
  #gb = None
  gb.printAll()
  gb = None
stuff()
```
